```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_csv_path = (r"C:\Users\yenia\Desktop\JHU_Bootcamp\Homework\Pandas-Challenge\PyCitySchools\Resources\schools_complete.csv")
student_csv_path = (r"C:\Users\yenia\Desktop\JHU_Bootcamp\Homework\Pandas-Challenge\PyCitySchools\Resources\students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_csv_path)
student_data_df = pd.read_csv(student_csv_path)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df
```

```python
#Group by school name so that duplicates are not counted
school_names_df = school_data_complete_df.groupby("school_name")
school_names_df.count()
```

```python
#Determine the total number of schools in the grouped data frame
total_schools = len(school_names_df)
print(total_schools)
```

```python
#Calculate the total number of students in the complete data fram
total_students = school_data_complete_df["Student ID"].count()
print(total_students)
```

```python
#Group by school name, extracting "budget" column
school_group_df = school_data_complete_df.groupby("school_name")["budget"]
school_budgets_df = school_group_df.max()
school_budgets_df
```

```python
#Calculate the total budget of all schools
total_budget = school_budgets_df.sum()
school_budgets_df.sum()
```

```python
#Calculate the average math score
avg_math_score = school_data_complete_df["math_score"].mean()
school_data_complete_df["math_score"].mean()
```

```python
#Calculate the average reading score
avg_reading_score = school_data_complete_df["reading_score"].mean()
school_data_complete_df["reading_score"].mean()
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math_df = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70,:]
passing_math_df
```

```python
#Determine the total number of students passing math
num_students_passing_math = len(passing_math_df)
print(len(passing_math_df))
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
percent_passing_math = (num_students_passing_math/total_students) * 100
print(percent_passing_math)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_df = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70,:]
passing_reading_df
```

```python
#Determine the total number of students passing reading
num_students_passing_reading = len(passing_reading_df)
print(len(passing_reading_df))
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
percent_passing_reading = (num_students_passing_reading/total_students) * 100
print(percent_passing_reading)
```

```python
#Merge dataframes of students passing each of the subjects into one
passing_both_df = pd.merge(passing_math_df,passing_reading_df)
passing_both_df
```

```python
#Determine the total number of students passing both subjects
num_students_passing_both = len(passing_both_df)
print(len(passing_both_df))
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
percent_passing_both = (num_students_passing_both/total_students) * 100
print(percent_passing_both)
```

```python
#Create data frame for results
raw_data_items = {

    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [percent_passing_math],
    "% Passing Reading": [percent_passing_reading],
    "% Overall Passing": [percent_passing_both],
    
    }
district_summary_results_df =pd.DataFrame(raw_data_items, columns=[       
    "Total Schools","Total Students","Total Budget","Average Math Score", "Average Reading Score","% Passing Math", "% Passing Reading","% Overall Passing"])
district_summary_results_df
```
